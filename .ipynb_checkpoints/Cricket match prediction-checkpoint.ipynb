### Import required packages

```python
import pandas as pd
import numpy as np

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import confusion_matrix, classification_report
```

### Read match data from a CSV file

```python
df = pd.read_csv("match_data.csv")
df
```

### Create numerical values for venue and time

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df["venue"] = encoder.fit_transform(df["venue"])
df["time"] = encoder.fit_transform(df["time"])
df.head()
```

### Extracting "team1" from "Match" column 

```python
team1 = []
for items in df["Match"]:
    team1.append(items.split()[0])
team1
```

```python
# adding the team1 column in the dataframe
df["team1"] = team1
```

### Create numerical values for toss-win, bat-first, and won columns

```python

toss_win_numerical  = np.where(df["team1"] == df["toss-win"],0,1)
df["toss_win_numerical"] = toss_win_numerical

bat_first_numerical = np.where(df["team1"] == df["bat_first"],0,1)
df["bat_first_numerical"] = bat_first_numerical

won_numerical = np.where(df["team1"] == df["won"],0,1)
df["won_numerical"] = won_numerical

df
```

```python
# dataframe with numerical features only
prediction_df = df[["venue","time","toss_win_numerical","bat_first_numerical","won_numerical"]]
prediction_df
```

```python
# creating input features
X = prediction_df.drop("won_numerical",axis=1)
X
```

```python
# output feature
Y = prediction_df["won_numerical"]
Y
```

```python
# splitting the data into training and testing data

X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
print(len(X_train),len(X_test))
```

### Train Decision Tree Classifier

```python
dtree = DecisionTreeClassifier(criterion="entropy")
dtree.fit(X_train,y_train)
```

### Predicting the output using the trained model

```python
import gradio as gr
```

```python
def venueNumerical(venue):
    if venue == "abu dhabi":
        return 0
    elif venue == "dubai":
        return 1
    else:
        return 2

def timeNumerical(time):
    if time == "day":
        return 0
    else:
        return 1

def teamNumerical(team):
    if team == "team1":
        return 0
    else:
        return 1
    
def teamString(value):
    if value == 0:
        return "team1"
    else:
        return "team2"
```

```python


def matchPrediction(venue,time,toss_win,bat_first):
    venue_numerical = venueNumerical(venue)
    time_numerical = timeNumerical(time)
    toss_win_numerical = teamNumerical(toss_win)
    bat_first_numerical = teamNumerical(bat_first)
    
    match_data = np.array([[venue_numerical,time_numerical,toss_win_numerical,bat_first_numerical]])
    
    prediction = dtree.predict(match_data)
    
    return teamString(prediction[0])
    
```

```python
# using Gradio library to develop a simple GUI
interface = gr.Interface(
    fn = matchPrediction,
    inputs = [gr.inputs.Dropdown(["dubai","abu dhabi","sharjah"], type="value", default="dubai", label="Venue"),
             gr.inputs.Dropdown(["day","night"], type="value", default="night", label="Time"),
             gr.inputs.Dropdown(["team1","team2"], type="value", default="team1", label="Toss Win"),
             gr.inputs.Dropdown(["team1","team2"], type="value", default="team1", label="Bat first"),
             ],
    outputs = ['text']
)
```

```python
interface.launch()
```

### Checking the accuracy of the model

```python
dtree.score(X_test,y_test)
y_pred = dtree.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print('Accuracy of decision tree classifier on test set: {:.4f}'.format(dtree.score(X_test, y_test)))
```
